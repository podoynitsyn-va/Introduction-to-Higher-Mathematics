1). Напишите код, который будет переводить полярные координаты в декартовы.

```python
def polar_to_dekart(p, fi)
    x = p * math.cos(fi)
    y = p * math.sin(fi)
    return [x,y]
```

2).	Напишите код, который будет рисовать график окружности в полярных координатах.

```python
def circle_function(radians, radius):
    return {"x": math.cos(radians) * radius,
            "y": math.sin(radians) * radius}
plt.polar(circle_function(2,10).get('x'), circle_function(2,10).get('y'))
```
