2)	Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
def vector_length(vector):
    value = 0
    for i in vector:
        value += i**2
    return math.sqrt(value) 

a = np.array([3,4])
b = np.array([3,4,5])
print('Длина вектора a = ' + str(vector_length(a)))
print('Длина вектора b = ' + str(vector_length(b)))
```

#### Задание 2

Почему прямые не кажутся перпендикулярными? 

```python
x = np.linspace(-5, 5, 21)
y = 3*x+1
y2 = (-1/3)*x+1
plt.plot(x,y)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
```

Прямые не кажутся перпендикулярными потому, что во-первых, установлены разные пределы отображения осей для X и для Y, во-вторых, размер фигуры, в которой мы производим отрисовку графика, не является одинаковым для X и для Y.
Установим размер фигуры и предел для оси 0Y такой же, как и для оси 0X:

```python
plt.figure(figsize = (8, 8))
x = np.linspace(-5, 5, 21)
y = 3*x+1
y2 = (-1/3)*x+1
plt.plot(x,y)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-5,5)
```

Теперь прямые кажутся перпендикулярными.

#### Задание 3

Напишите код на Python, реализующий построение графиков:
1.	окружности,
2.	эллипса,
3.	гиперболы.


1. Окружность:

```python
plt.figure(figsize = (6, 6))
x = np.linspace(-3.0, 10.0, 100)
y = np.linspace(-3.0, 10.0, 100)
X, Y = np.meshgrid(x,y)
F = (X-4)**2 + (Y-2)**2 - 16
plt.contour(X,Y,F,[0])
plt.grid(linestyle="-", color='grey')
print("Центр окружности - точка [4,2], радиус 4")
plt.show()
```

2. Эллипс

```python
plt.figure(figsize = (8, 8))
x = np.linspace(0, 8, 100)
y = np.linspace(0, 8, 100)
X, Y = np.meshgrid(x,y)
F = ((X-4)**2)/9 + ((Y-4)**2)/4 - 1
c = math.sqrt(9-4)+4
c1 = -math.sqrt(9-4)+4
plt.plot([c,c1],[4,4], marker='o', color='r')
plt.contour(X,Y,F,[0])
plt.grid(color='lightgray',linestyle='--')
print(f"Центр эллипса - точка [4,4], фокусы = точки [{c1},4] и [{c},4] ")
plt.show()
```

3. Гипербола

```python
plt.figure(figsize = (8, 8))
x = np.linspace(-5, 13, 100)
y = np.linspace(-5, 13, 100)
X, Y = np.meshgrid(x,y)
F = ((X-4)**2)/9 -((Y-4)**2)/4 - 1
c = math.sqrt(9+4)+4
c1 = -math.sqrt(9+4)+4
plt.plot([c,c1],[4,4], marker='o', color='r')
asimpt1 = 2/3*(x-4)+4
asimpt2 = -2/3*(x-4)+4
plt.plot(x,asimpt1)
plt.plot(x,asimpt2)
plt.contour(X,Y,F,[0])
plt.grid(color='lightgray',linestyle='--')
print(f"фокусы гиперболы = точки [{c1},4] и [{c},4] ")

plt.show()
```
